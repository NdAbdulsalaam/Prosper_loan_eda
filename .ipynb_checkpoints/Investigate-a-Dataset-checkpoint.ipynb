## Introduction

> What can we say about the success of a movie before it is released? Are there certain companies (Pixar?) that have found a consistent formula? Given that major films costing over $100 million to produce can still flop, this question is more important than ever to the industry. Film aficionados might have different interests. Can we predict which films will be highly rated, whether or not they are a commercial success?

> This data set contains information about 10,000 movies collected from The Movie Database (TMDb), including user ratings and revenue.

> The dataset contains the following features:-
- id - A unique identifier for each movie.
- cast - The name of lead and supporting actors.
- crew - The name of Director, Editor, Composer, Writer etc.
- budget - The budget in which the movie was made.
- genre - The genre of the movie, Action, Comedy ,Thriller etc.
- homepage - A link to the homepage of the movie.
- original_title - The title of the movie before translation or adaptation.
- overview - A brief description of the movie.
- popularity - A numeric quantity specifying the movie popularity.
- production_companies - The production house of the movie.
- release_date - The date on which it was released.
- revenue - The worldwide revenue generated by the movie.
- runtime - The running time of the movie in minutes.
- tagline - Movie's tagline.
- title - Title of the movie.
- vote_count - the count of votes recieved.
- release_year = the year the movie was released
- vote_average - average ratings the movie recieved.

## Data Cleaning and Engineering

## Fix incorrect datatypes:

```python
df.info()
```

## Rename `original_title` as `title`

```python
df.rename(columns= {'original_title': 'title'}, inplace= True)
```

## Change release_date to datetime and create the month and day columns for easy analysis

```python
## Convert to datetime
df['release_date'] = pd.to_datetime(df.release_date)

## Create release_month and release date name
df['release_month'] = df['release_date'].dt.month_name()
df['release_day'] = df['release_date'].dt.day_name()

## Drop the release_date column
df.drop(columns=['release_date'], inplace = True)

## Confirm changes
df.head(2)
```

## Create the `profit` column from `revenue` and `budget` and drop them
> Since profit is the sole of business

```python
df['profit'] = df.revenue - df.budget
df.drop(columns= ['revenue', 'budget'], inplace = True)

## Confirm changes
df.head(3)
```

## Re-write the keywords and genres columns
> This will made the word in the columns readable

```python
df['keywords'] = df['keywords'].apply(lambda x: x.replace("|", " "))
df['cast'] = df['cast'].apply(lambda x: x.replace("|", ", "))
df['genres'] = df['genres'].apply(lambda x: x.replace("|", ", "))
df['production_companies'] = df['production_companies'].apply(lambda x: x.replace("|", ", "))

## Confirm changes
df.head(3)
```

## Categorizing the runtime using `pandas cut` into
- Short
- Moderate
- Long
- Very long

**Why:**<br>
This values are so many that it cannot be used for analysis (no meaniful inference can bo gotten) unlesss categorized

```python
df.runtime.nunique()
```

```python
df.runtime.describe()
```

```python
bin_edges = [0, 30, 60, 120, 705]
bin_names = ['Short', 'Moderate', 'Long', 'Very long']
df['duration'] = pd.cut(df.runtime, bin_edges, labels = bin_names)

## Confirm chages
df.head(3)
```

## Exploratory Data Analysis

## Define function for resuability

```python
def barhplot(x, y, xlabel, title):
    plt.figure(figsize=  (14, 8))
    plt.barh(x, y, align = 'center')
    plt.gca().invert_yaxis()
    plt.xlabel(xlabel, fontsize = 18)
    plt.title(title, fontsize = 18)
    plt.show();

def barplot(x, y, xlabel, title):
    plt.figure(figsize=  (10, 7))
    plt.bar(x, y, align = 'center')
    plt.xlabel(xlabel, fontsize = 18)
    plt.title(title, fontsize = 18)
    plt.show();

def lineplot(value, xlabel, ylabel, title):
    plt.figure(figsize=  (17, 9))
    plt.plot(value)
    plt.xlabel(xlabel, fontsize = 18)
    plt.ylabel(ylabel, fontsize = 18)
    plt.title(title, fontsize = 23)
    plt.show();

def sortby(column):
    df1 = df.sort_values(by = column, ascending = False)
    return df1
```

## Q5. Checking the influence of duration on popularity and profit

```python
duration_pop = df.groupby(['duration']).mean()['popularity']
duration_pop.sort_values(ascending= False)
```

```python
barplot(duration_pop.index, duration_pop, "Popularity", "Popularity of movies based on the duration")
```

supprisingly, on average, the longer the movie, the more popular it is.
Then, we need to check if profit also follow thesame trend

```python
duration_prof = df.groupby(['duration']).mean()['profit']
duration_pop.sort_values(ascending= False)
```

```python
barplot(duration_prof.index, duration_prof, "Profit", "Profit from movies based on the duration")
```

That awesom, the profit also increases as the duration increasea and perphaps, with increase in popularity

## Q6. Checking the influence of the day a movie was on popularity and profit
> To check on which days are the popular moviese rerleased so that they can be target for future release

```python
day_pop = df.groupby(['release_day']).mean()['popularity']
day_pop.sort_values(ascending= False)
```

## Q7. Checking the influence of the month a movie was released on popularity and profit
> To check on which months are the popular moviese rerleased so that they can be target for future release

```python
month_pop = df.groupby(['release_month']).mean()['popularity'].sort_values(ascending= False)
month_pop
```

From this graph, it'll be a good advice to that a movie is released in June, December or at least May.
